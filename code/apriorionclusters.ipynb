```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('C:/Users/efe44/Desktop/veri madenciliği proje/Clustered_OnlineRetail.xlsx')
```

```python
cluster_id = 2
cluster_data = df[df['Cluster'] == cluster_id]
```

```python
invoice_products = cluster_data.groupby('InvoiceNo')['StockCode'].apply(list)
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(invoice_products)
one_hot = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(one_hot, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5, num_itemsets=len(frequent_itemsets))
```
