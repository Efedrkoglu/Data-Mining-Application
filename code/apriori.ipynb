```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('C:/Users/efe44/Desktop/veri madenciliği proje/OnlineRetail.xlsx')
```

```python
print(df.info())
```

```python
df['StockCode'] = df['StockCode'].astype(str)
transactions = df.groupby('InvoiceNo')['StockCode'].apply(list)
transactions
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
transaction_df = pd.DataFrame(te_ary, columns=te.columns_)
transaction_df
```

```python
frequent_items = apriori(transaction_df, min_support=0.01, use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items, num_itemsets=1, metric="confidence", min_threshold=0.8)
rules
```
